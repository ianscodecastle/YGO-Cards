```python
import requests
import pandas as pd
import json
```

```python
url = 'https://db.ygoprodeck.com/api/v7/cardinfo.php'
spellcasters = requests.get(url=url, params={'race':'Spellcaster'})
spellcasters_json = spellcasters.json()['data']
```

```python
spellcasters_df = pd.DataFrame(spellcasters_json)
spellcasters_prices = pd.json_normalize(spellcasters_df['card_prices'].explode(0))
display(spellcasters_prices)
```
